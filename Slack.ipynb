```python
#Question_1
import math
def even(number):
    if(number%2==0):
        return True
    else:
        return False
l=list(range(1,21))
even_sum=0
odd_prod=1
mean=sum(l)/len(l)
if (len(l)%2==0):
    median=(l[(len(l)//2)-1]+l[len(l)//2])/2
else:
    median=l[(len(l)//2)-1]
x,y=0,0
for items in l:
    if even(items):
        even_sum+=items
    else:
        odd_prod*=items
    y=math.fabs(items-mean)
    y=y*y
    x+=y
z=x/len(l)
standard_deviation=math.sqrt(z)
print(even_sum)
print(odd_prod)
print(mean)
print(median)
print(standard_deviation)
```

```python
#Question_3
import matplotlib.pyplot as plt
def Fibonacci(n):
    l=[]
    while len(l)<n:
        if(len(l)==0):
            l.append(0)
            continue
        if(len(l)==1):
            l.append(1)
            continue
        l.append(l[-1]+l[-2])
    else:
        return l
n=int(input())
Fibonacci_List=(Fibonacci(n))
print(Fibonacci_List)
numbers=list(range(n+1))
numbers.pop(0)
plt.xlabel("No of terms")
plt.ylabel("Fibonacci")
plt.plot(numbers,Fibonacci_List)
plt.show()
```
